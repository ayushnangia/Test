```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertForSequenceClassification, DistilBertTokenizer
model_n='distilbert-base-uncased-finetuned-sst-2-english'
```

```python
pt_model=AutoModelForSequenceClassification.from_pretrained(model_n)
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_n)
```

```python
input=tokenizer("Hi there bro")
```

```python
input
```

```python
sentec_neg="Hate english"
sentec_pos="Math is adore"
```

```python
pt_batch=tokenizer(
    [sentec_pos,sentec_neg],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="pt"
)
```

```python
pt_batch.attention_mask
```

```python
pt_outputs=pt_model(**pt_batch)
```

```python
pt_outputs
```

```python
from torch import nn
pt_predictions=nn.functional.softmax(pt_outputs.logits,dim=-1)
```
